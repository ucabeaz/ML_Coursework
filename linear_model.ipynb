```python
import numpy as np 
import matplotlib.pyplot as plt  
import pandas as pd 
```

```python
df = pd.read_csv('cleaned_dataset.csv') 
```

```python
df.groupby('Sector')['Price'].mean().plot(kind='bar')
```

```python
# Sort by ticker and year
df = df.sort_values(['Ticker', 'Year']).copy()

# Create next-year return target
df['target_next_year_return'] = df.groupby('Ticker')['Stock_Return'].shift(-1)

# Remove rows where next year's return is unavailable
df_model = df.dropna(subset=['target_next_year_return']).copy()

print(df_model.shape)
print(df_model[['Ticker', 'Year', 'Stock_Return', 'target_next_year_return']].head(10))
```
